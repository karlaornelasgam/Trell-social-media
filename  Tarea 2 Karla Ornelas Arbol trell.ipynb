```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

# Dividir los datos en características (X) y la variable objetivo (y)
X = train.drop('age_group', axis=1)
y = train['age_group']

# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=60)

# Instanciar el clasificador de árbol de decisión
model = DecisionTreeClassifier()

# Entrenar el modelo con los datos de entrenamiento
model.fit(X_train, y_train)

# Hacer predicciones en los datos de prueba
y_pred = model.predict(X_test)

# Calcular la precisión y F1 Score del modelo
accuracy = accuracy_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred, average='weighted')
conf_matrix = confusion_matrix(y_test, y_pred)

print("Accuracy:", accuracy)
print("F1 Score:", f1)
print("Confusion Matrix:")
print(conf_matrix)

# También puedes visualizar el árbol de decisión si lo deseas
#plt.figure(figsize=(15,10))
#plot_tree(model, filled=True, feature_names=X_train.columns, class_names=['1', '2', '3', '4'])
#plt.show()
# el kernel nunca cargo
```
